```python
def get_div_content(div):
    """
    Fonction récursive pour récupérer le contenu d'une division et de ses sous-divisions.
    """
    div_content = div.text.strip()
    
    # Parcourir les sous-divisions récursivement
    for sub_div in div.find_all("div"):
        div_content += "\n" + get_div_content(sub_div)

    return div_content

# Effectuer la requête HTTP pour obtenir le contenu de la page

response = requests.get(url)
content = response.content

# Analyser le contenu HTML avec BeautifulSoup
soup = BeautifulSoup(content, "html.parser")

# Trouver toutes les div dans le contenu HTML
divs = soup.find_all("div")

# Récupérer le contenu de chaque division et de ses sous-divisions
for div in divs:
    div_content = get_div_content(div)
    print(div_content)
```

```python
def scroll_to_bottom(driver):
    """
    Faire défiler la page jusqu'en bas.
    """
    # Utiliser la touche End pour faire défiler jusqu'en bas
    for i in range(10):
        driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)
        time.sleep(1)  # Attendre pour permettre le chargement du contenu
    #driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)
    #time.sleep(4)  # Attendre pour permettre le chargement du contenu
    #driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)

    # Vérifier si la page a atteint le bas
    scroll_height = driver.execute_script("return document.documentElement.scrollHeight")
    window_height = driver.execute_script("return window.innerHeight")
    print(scroll_height, window_height)
    if scroll_height == window_height:
        return False

    return True
```
